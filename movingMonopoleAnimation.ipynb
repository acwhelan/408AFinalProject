```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation

# Parameters
num_vectors = 30
vector_length = 2
charge_position = np.array([0, 0])
num_frames = 100  # Number of frames in the animation

# Create a grid of points
x = np.linspace(-5, 5, num_vectors)
y = np.linspace(-5, 5, num_vectors)
X, Y = np.meshgrid(x, y)

# Function to update the vectors
def update_quiver(num, Q, X, Y):
    """ Updates the vector field for animation """
    U = (X - charge_position[0]) / np.sqrt((X - charge_position[0])**2 + (Y - charge_position[1])**2)
    V = (Y - charge_position[1]) / np.sqrt((X - charge_position[0])**2 + (Y - charge_position[1])**2)
    
    # Scale vectors to create an outward moving effect
    scale_factor = 1 + 0.05 * num
    U *= vector_length * scale_factor
    V *= vector_length * scale_factor
    
    Q.set_UVC(U, V)
    return Q,

# Create a figure and axis
fig, ax = plt.subplots()
ax.set_xlim(-5, 5)
ax.set_ylim(-5, 5)

# Create quiver plot
Q = ax.quiver(X, Y, X, Y, scale=1)

# Store the animation in a global variable
anim = animation.FuncAnimation(fig, update_quiver, fargs=(Q, X, Y),
                               frames=num_frames, interval=50, blit=False)

# Display the animation
plt.show()
```
